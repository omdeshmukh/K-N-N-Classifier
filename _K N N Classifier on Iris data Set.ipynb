## Scikit Learn

```python
#Load the IRIS dataset from the Scikit learn library

from sklearn.datasets import load_iris
iris=load_iris()
```

```python
#Store the feature matrix i.e. X and Response vector Y

X=iris.data
y=iris.target
```

```python
# Store the feature names and target data

feature_names=iris.feature_names #['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
target_data=iris.target_names   #['setosa', 'versicolor', 'virginica']
```

```python
# Print the feature Names and target names of our data

print("feature Names ", feature_names,'\n')
print("Target Data   " , target_data)
```

```python
# check the type of X and y

print("Type of X is ", type(X)," and Type of y is :", type(y))
```

```python
# print the first 5 rows of an array

print("The top 5 rows of data :","\n", X[0:5])
```

```python
# splitting X and y into training and testing sets 
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1) 
```

```python
print('X Train Size',X_train.size)

print('X Test Size',X_test.size)
```

```python
# printing the shapes of the new X objects 
print(X_train.shape) 
print(X_test.shape)  
```

```python
# printing the shapes of the new y objects abs
print(y_train.shape) 
print(y_test.shape)
```

```python
# training the model on training set 
from sklearn.neighbors import KNeighborsClassifier 
knn = KNeighborsClassifier(n_neighbors=3) 
knn.fit(X_train, y_train)
```

```python
import pandas as pd

test_df=pd.DataFrame(X_test,)
test_df['labels']=y_test
print(test_df)

```

```python
# making predictions on the testing set 
y_pred = knn.predict(X_test) 
```

```python
test_df['y_pred']=y_pred
```

```python
test_df
```

```python
# comparing actual response values (y_test) with predicted response values (y_pred) 
from sklearn import metrics 
print("kNN model accuracy:", (metrics.accuracy_score(y_test, y_pred)*100))
```

```python
# making prediction for out of sample data 
sample = [[6.7,3.0,5.2,2.3], [4.3,3.0,1.1,0.1]] 
preds = knn.predict(sample) 
preds
pred_species = [iris.target_names[p] for p in preds] 
print("Predictions:", pred_species) 
```

```python
# saving the model 
from sklearn.externals import joblib 
joblib.dump(knn, 'iris_knn.pkl')
```
